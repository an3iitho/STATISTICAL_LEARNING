# Problema #1

## Algoritmo de Canny, Operador de Sobel y Operador Laplaciano

```python
def alg_deteccion_bordes(dir_imagen, threshold1 = 100, threshold2 = 200):
    img = cv2.imread(dir_imagen)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    ## Algoritmo de Canny
    edges = cv2.Canny(img, threshold1, threshold2)
    nombre_img_canny = dir_imagen.replace('.jpg','') + '-canny.jpg'
    cv2.imwrite(nombre_img_canny, edges)
    
    ## Operador de Sobel
    sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
    sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

    grad = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
    nombre_img_sobel = dir_imagen.replace('.jpg','') + '-sobel.jpg'
    cv2.imwrite(nombre_img_sobel, grad)
    
    ## Operador Laplaciano
    laplaciano = cv2.Laplacian(img, cv2.CV_16S, ksize = 3)
    abs_dst = cv2.convertScaleAbs(laplaciano)
    nombre_img_laplaciano = dir_imagen.replace('.jpg','') + '-laplaciano.jpg'
    cv2.imwrite(nombre_img_laplaciano, abs_dst)
    
    return nombre_img_canny, nombre_img_sobel, nombre_img_laplaciano
```

```python
#Imagen original
verImagen("antigua.jpg")
```

### Algoritno de Canny

```python
canny, sobel, laplaciano = alg_deteccion_bordes("antigua.jpg")
verImagen(canny)
```

## Comparativa 

|**Algoritmo**|**Ventajas**|**Desventajas**|
|:-:|:-:|:-:|
|**Algoritmo de Canny**|Altamente adaptable|Puede tener problemas con el ruido en algunas imágenes|
||Puede utilizar Sobel como uno de sus pasos|Computacionalmente es más costoso|
||Alto nivel de exactitud en la imágen resultante||
|**Operador de Sobel**|Computacionalmente menos costoso|Tiende a fallar en bordes débiles|
||Al ser un filtro es más sencillo|Tiende sobre estimar zonas de poco contraste|
||Mejor en presencia de ruido||
|**Operador Laplaciano**|Filtro computacionalmente menos costoso|La imagen resultante no es tan exacta|
|||Tiende a generar imagenes ruidosas|

## Problema #2

```python
def bordes(dir_imagen, threshold1 = 100, threshold2 = 200):
    img = cv2.imread(dir_imagen)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    ## Algoritmo de Canny
    edges = cv2.Canny(img, threshold1, threshold2)
    nombre_img_canny = dir_imagen.replace('.jpg','') + '-bordes.jpg'
    cv2.imwrite(nombre_img_canny, edges)
    
    return nombre_img_canny
```

```python
verImagen(bordes("lago.jpg"))
```

```python
def contornos(dir_imagen):
    img = cv2.imread(dir_imagen)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    #Transforma la imagen a binaria
    _, binaria = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)

    #Encuentra los contornos de la imagen en binaria
    contours, hierarchy = cv2.findContours(binaria, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    #Dibuja contornos en la imagen original, marcados en rojo
    img_resultante = cv2.drawContours(img, contours, -1, (0, 0, 255), 2)
    
    nombre_img = dir_imagen.replace('.jpg','') + '-contornos.jpg'
    cv2.imwrite(nombre_img, img_resultante)

    return nombre_img

verImagen(contornos("lago.jpg"))
```

### Diferencias entre bordes y contornos en OpenCV

- Bordes: son puntos que tienen un cambio de contraste grande entre unos y otros, visto en términos de ecuaciones, los bordes representan puntos máximos y mínimos locales en ciertas partes de la imagen. Este enfoque puede ser utilizado para identificar varios contrastes dentro de un objeto, identificando el detalle completo en las zonas más importantes. 

- Contornos: son uniones de puntos (o bordes) que forman líneas cerradas de grandes porciones de cambios en la función, con lo que se forman contornos que pueden representar objetos en la función. Este enfoque puede ser utilizado cuando el detalle central a los bordes no interesa, sino únicamente las formas dentro de la imagen. 


## Problema #3

```python
def transformada_Hough(dir_imagen, theta = 50, minLinLength = 5, maxLineGap = 10):
    img = cv2.imread(dir_imagen)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    #Usando Canny para detectar los bordes
    dst = cv2.Canny(gray, 50, 200, None, 3)
    
    #Imagen de bordes agregando BGR para mostrar las lineas
    cdst = cv2.cvtColor(dst, cv2.COLOR_GRAY2BGR)
    
    #Parametros:
        #rho : Resolucion inicial del parametro, usando 1 pixel. 
        #theta: Resolucion de θ en redianes. 
        #threshold: Minimo numero de intersecciones para detectar una linea
        #minLinLength: Minimo numero de puntos que forman una linea
        #maxLineGap: Distancia maxima entre puntos para ser considerados de una misma linea
    
    lineas = cv2.HoughLinesP(dst, 1, np.pi / 180, theta, None, minLinLength, maxLineGap)
    
    #Dibujar las lineas encontradas de color rojo en la imagen    
    if lineas is not None:
        for i in range(0, len(lineas)):
            l = lineas[i][0]
            cv2.line(cdst, (l[0], l[1]), (l[2], l[3]), (0,0,255), 3, cv2.LINE_AA)
    
    #Gurdando la imagen
    nombre_img = dir_imagen.replace('.jpg','') + '-transformada_Hough.jpg'
    cv2.imwrite(nombre_img, cdst)

    return nombre_img

verImagen(transformada_Hough("campo1.jpg"))
```

```python
verImagen(transformada_Hough("campo2.jpg", 70, 40, 10))
```
