```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('C:/Users/alvar/Pictures/img_prueba.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize = (80,8))
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2.imwrite('img_prueba_gray.jpg', gray)
```

```python
plt.imshow(gray, cmap = 'gray')
```

```python
color = ('b', 'g', 'r')

for i, c in enumerate(color):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0, 256])
    
plt.show()
```

```python
ancho = image.shape[0]
alto = image.shape[1]
print("Ancho: ", ancho, " Alto:", alto)

gray = np.zeros((ancho, alto, 1))

for i in range(0, ancho):
    for j in range(0, alto):
        pixel = image[i, j]
        #BGR = Azul, Verde y Rojo
        #ponderado = (int(0.114 * pixel[0]) + int(0.587 * pixel[1]) + int(0.299 * pixel[2])) para BGR
        ponderado = (int(0.299 * pixel[0]) + int(0.587 * pixel[1]) + int(0.114 * pixel[2])) #para RGB
        gray[i, j] = ponderado

cv2.imwrite('img_prueba_gray2.jpg', gray)
```

```python
image2 = cv2.imread('img_prueba_gray2.jpg', cv2.IMREAD_COLOR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
```

```python
#Aritmetico

for i in range(0, ancho):
    for j in range(0, alto):
        pixel = image[i, j]
        aritmetico = int((int(pixel[0]) + int(pixel[1]) + int(pixel[2]))/3) 
        gray[i, j] = aritmetico

cv2.imwrite('img_prueba_gray3.jpg', gray)
```

```python
image2 = cv2.imread('img_prueba_gray3.jpg', cv2.IMREAD_COLOR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
```

## Escala de grises 3D

```python
gris_rojo = np.zeros((ancho, alto, 1))
gris_verde = np.zeros((ancho, alto, 1))
gris_azul = np.zeros((ancho, alto, 1))

for i in range(0, ancho):
    for j in range(0, alto):
        pixel = image[i, j]
        gris_rojo[i,j] = pixel[0]
        gris_verde[i,j] = pixel[1]
        gris_azul[i,j] = pixel[2]
        
cv2.imwrite('gris_rojo.jpg', gris_rojo)
cv2.imwrite('gris_verde.jpg', gris_verde)
cv2.imwrite('gris_azul.jpg', gris_azul)
```

```python
rojo = cv2.imread('gris_rojo.jpg', cv2.IMREAD_COLOR)
verde = cv2.imread('gris_verde.jpg', cv2.IMREAD_COLOR)
azul = cv2.imread('gris_azul.jpg', cv2.IMREAD_COLOR)

rojo = cv2.cvtColor(rojo, cv2.COLOR_BGR2RGB)
verde = cv2.cvtColor(verde, cv2.COLOR_BGR2RGB)
azul = cv2.cvtColor(azul, cv2.COLOR_BGR2RGB)
```
