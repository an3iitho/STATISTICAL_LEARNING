### Problema #1

```python
def coloresActivos(dir_imagen, color):
    #Importar la imagen
    img = cv2.imread(dir_imagen)
    ancho = img.shape[0]
    alto = img.shape[1]
    print("Ancho: ", ancho, " Alto:", alto)
    azul = 0
    verde = 0
    rojo = 0
    
    nueva_img = np.zeros((ancho, alto, 3))
    
    if color == 1:#Unicamente color azul
        azul = 1
    elif color == 2: #Unicamente color verde
        verde = 1
    elif color == 3: #Unicamente color rojo
        rojo = 1
    elif color == 10: #Unicamente color rojo y verde
        rojo = 1
        verde = 1
    elif color == 20: #Unicamente color verde y azul
        verde = 1
        azul = 1
    elif color == 30: #Unicamente color azul y rojo
        azul = 1
        rojo = 1
        
    print("rojo ", rojo, " azul ", azul, " verde ", verde)

    for i in range(0, ancho):
        for j in range(0, alto):
            pixel = img[i, j]
            #BGR = Azul, Verde y Rojo
            nueva_img[i, j, 0] = int(azul * pixel[0])
            nueva_img[i, j, 1] = int(verde * pixel[1])
            nueva_img[i, j, 2] = int(rojo * pixel[2])
            
    nombre_img = "nueva_img-" + str(color) + ".jpg"
    
    cv2.imwrite(nombre_img, nueva_img)
    
    return nombre_img

def verImagen(dir_imagen):
    image2 = cv2.imread(dir_imagen, cv2.IMREAD_COLOR)
    image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
    plt.imshow(image2)
```

### Problema #2

```python
def convierte_bn_a_color(img_bn_rojo, img_bn_verde, img_bn_azul, nombre_img):
    #Importar la imagen
    img_rojo = cv2.imread(img_bn_rojo, 0)
    img_verde = cv2.imread(img_bn_verde, 0)
    img_azul = cv2.imread(img_bn_azul, 0)
    ancho = img_rojo.shape[0]
    alto = img_rojo.shape[1]
    print("Ancho: ", ancho, " Alto:", alto)
    
    nueva_img = np.zeros((ancho, alto, 3))

    for i in range(0, ancho):
        for j in range(0, alto):
            pixel_a = img_azul[i, j]
            pixel_v = img_verde[i, j]
            pixel_r = img_rojo[i, j]
            ##BGR = Azul, Verde y Rojo
            nueva_img[i, j, 0] = int(pixel_a / 0.299)
            nueva_img[i, j, 1] = int(pixel_v / 0.587)
            nueva_img[i, j, 2] = int(pixel_r / 0.114)
            
    cv2.imwrite(nombre_img, nueva_img)
```

```python
def convierte_a_bn_3d(img_color):
    #Importar la imagen
    img = cv2.imread(img_color)
    
    ancho = img.shape[0]
    alto = img.shape[1]
    print("Ancho: ", ancho, " Alto:", alto)
    
    ##BGR = Azul, Verde y Rojo
    n_gris_azul = np.zeros((ancho, alto, 1))
    n_gris_verde = np.zeros((ancho, alto, 1))
    n_gris_rojo = np.zeros((ancho, alto, 1))

    for i in range(0, ancho):
        for j in range(0, alto):
            pixel = img[i, j]
            n_gris_azul[i,j] = pixel[0]
            n_gris_verde[i,j] = pixel[1]
            n_gris_rojo[i,j] = pixel[2]

    nombre_img_salida = img_color
    cv2.imwrite('nueva_prueba_gris_azul.jpg', n_gris_azul)
    cv2.imwrite('nueva_prueba_gris_verde.jpg', n_gris_verde)
    cv2.imwrite('nueva_prueba_gris_rojo.jpg', n_gris_rojo)
    
```

### Problema #4

```python
def histograma_x_color(dir_img):
    #Importar la imagen
    img = cv2.imread(dir_img)
    #BGR = Azul, Verde y Rojo
    print(img.shape)
    azul = np.array(img[:, :, 0]).flatten()
    verde = np.array(img[:, :, 1]).flatten()
    rojo = np.array(img[:, :, 2]).flatten()
    
    #plt.figure(figsize=(10,6))
    plt.hist(azul, bins  = 256, alpha = 0.75, color = 'blue')
    plt.axvline(azul.mean(), color = 'k', linestyle = 'dashed', linewidth = 1)
    plt.title("Histograma color azul") 
    plt.show()
    plt.hist(verde, bins  = 256, alpha = 0.75, color = 'green')
    plt.axvline(verde.mean(), color = 'k', linestyle = 'dashed', linewidth = 1)
    plt.title("Histograma color verde") 
    plt.show()
    plt.hist(rojo, bins  = 256, alpha = 0.75, color = 'red')
    plt.axvline(rojo.mean(), color = 'k', linestyle = 'dashed', linewidth = 1)
    plt.title("Histograma color rojo") 
    plt.show()
    
    #Imagen a blanco y negro
    gray = np.multiply(azul, 0.114) + np.multiply(verde, 0.587) + np.multiply(rojo, 0.299)
    gray = gray.astype(np.int)
    plt.hist(gray, bins  = 256, alpha = 0.5, color = 'black')
    plt.axvline(gray.mean(), color = 'red', linestyle = 'dashed', linewidth = 1)
    plt.title("Histograma blanco y negro") 
    plt.show()
```

### Problema #5

#### La escala de grises ponderada 

Debido a que el ojo humano percibe distintas intesidades de luz en función del color que se observe, el cálculo para realizar imagenes equivalentes en blanco y negro (escala de grises o luminancia) debe realizarse como una media ponderada de las distintas componentes de color de cada pixel. La medida recomendada es 0.3 para rojo, 0.59 para verde y 0.11 para azul. La siguiente función crea imagenes en blanco y negro con los valores estándar, pero se pueden modificar según los paramtros dados a la función.

```python
def a_byn(dir_img, pon_rojo = 0.3, pon_verde = 0.59, pon_azul = 0.11):
    #Importar la imagen
    img = cv2.imread(dir_img)
    #BGR = Azul, Verde y Rojo
    azul = np.array(img[:, :, 0]).flatten()
    verde = np.array(img[:, :, 1]).flatten()
    rojo = np.array(img[:, :, 2]).flatten()

    #Imagen a blanco y negro
    gray = np.multiply(azul, pon_azul) + np.multiply(verde, pon_verde) + np.multiply(rojo, pon_rojo)
    gray = gray.astype(np.int)
    gray = gray.reshape((img.shape[0], img.shape[1]))
    
    nombre_img = dir_img.replace('.jpg','_byn_ponderado.jpg')
    cv2.imwrite(nombre_img, gray)
    
    return nombre_img
```

```python
#Usando la ponderación estándar
verImagen(a_byn("img_prueba3.jpg"))
```

```python
#Usando una ponderacion diferente
verImagen(a_byn("img_prueba3.jpg", 0.4, 0.1, 0.7))
```

### Problema #6

#### HSV

Es una representación alternativa de los colores al modelo RGB que es muy utilizado en la industria, creado con la finalidad de lograr una interpretación mas cercana a como los humanos percibe los colores. 

En este modelo, los colores se mapean en un espacio radial, alrededor de un eje central de colores neutros que varía desde negro en la parte inferior a blanco en la parte superior. De la siguiente forma:


<img src="https://handmap.github.io/images/posts/2016-10-26/opencv-hsv-cone.jpg"/>


* Tinte o matiz (Hue): Ángulo que representa el matiz, define la tonalidad del color.
* Saturación: Nivel saturación del color, dado entre 0 y 1, 0 representa sin saturación (o color blanco), hasta 1 que sería el matiz en toda su intensidad. 
* Brillo: Nivel del brillo entre 0 y 1. 0 es negro; 1, blanco. Al igual que la saturación puede darse en porcientos entre 0% y 100%. De esta forma el 50% indica el nivel medio o normal del brillo del color.

La ventaja de este modelo es que podemos elegir únicamente un tinte, y tenemos todas las saturación y brillos del color, con lo que es más fácil para un modelo de Machine Learning identificar elementos de ese color elegido, si estos elementos tienen un amplio rango de saturaciones y brillos.
