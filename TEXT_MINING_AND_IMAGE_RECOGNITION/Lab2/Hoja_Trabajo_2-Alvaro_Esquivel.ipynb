### Problema #1:
Desarrolle una función que permita hacer la binarización de una imagen, para realizarlo puede utilizar el enfoque que prefiera.

```python
def binarizacion(dir_img, umbral):
    #Importar la imagen
    img = cv2.imread(dir_img)
    
    #BGR = Azul, Verde y Rojo
    print(img.shape)
    azul = np.array(img[:, :, 0]).flatten()
    verde = np.array(img[:, :, 1]).flatten()
    rojo = np.array(img[:, :, 2]).flatten()
    
    #Transformacion a blanco y negro
    gray = np.multiply(azul, 0.114) + np.multiply(verde, 0.587) + np.multiply(rojo, 0.299)
    gray = gray.astype(np.int)
    #Binarizacion
    gray[gray > umbral] = 255
    gray[gray < umbral] = 0
    gray = gray.reshape((img.shape[0], img.shape[1]))
    
    nombre_img = dir_img.replace('.jpg','_binarizada.jpg')
    cv2.imwrite(nombre_img, gray)
    
    return nombre_img, gray

def verImagen(dir_imagen):
    image2 = cv2.imread(dir_imagen, cv2.IMREAD_COLOR)
    image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
    plt.imshow(image2)
```

```python
verImagen(binarizacion("img_prueba3.jpg", 100)[0])
```

## Problema #2:
Desarrolle una función que dada dos imágenes a color, realice las operaciones SUMA, RESTA, AND, OR , XOR 
con ambas imágenes recuerde que para realizar las operaciones lógicas es necesario pasar
la imagen a escala de grises y luego binarizarla.

```python
#operacion 1: suma 
#operacion 2: resta 
def operaciones(dir_img1, dir_img2, operacion):
    #Importar la imagen
    img1 = cv2.imread(dir_img1)
    img2 = cv2.imread(dir_img2)
    
    #BGR = Azul, Verde y Rojo
    alto1, ancho1, _ = img1.shape
    alto2, ancho2, _ = img2.shape
    print(img1.shape, img2.shape)
    print(max(alto1, alto2), max(ancho1, ancho2))
    
    #Crea matrices de 0 del tamaño de la imagen mayor
    zeros_img1 = np.zeros((max(alto1, alto2), max(ancho1, ancho2), 3))
    zeros_img2 = np.zeros((max(alto1, alto2), max(ancho1, ancho2), 3))
    
    #Agraga las imagenes originales a las matrices de ceros
    zeros_img1[:img1.shape[0],:img1.shape[1], :img1.shape[2]] = img1
    zeros_img2[:img2.shape[0],:img2.shape[1], :img2.shape[2]] = img2
    
    print(zeros_img1.shape, zeros_img2.shape)
    
    if (operacion == 1):
        #Suma de imagenes
        resultado = np.add(zeros_img1, zeros_img2)
        operacion = "suma"
    else:
        #Resta de imagenes
        resultado = np.subtract(zeros_img1, zeros_img2)
        operacion = "resta"
        
    nombre_img = dir_img1.replace('.jpg','') + '-' + dir_img2.replace('.jpg','')
    nombre_img = nombre_img + '-' + operacion + '.jpg'
    cv2.imwrite(nombre_img, resultado)
    
    return nombre_img
```

```python
#operacion 1: AND 
#operacion 2: OR 
#operacion 3: XOR

def operaciones_logicas(dir_img1, dir_img2, umbral, operacion):
    #Importar la imagen
    img1 = binarizacion(dir_img1, umbral)[1]
    img1[img1 > 0] = 1
    
    img2 = binarizacion(dir_img2, umbral)[1]
    img2[img2 > 0] = 1
    
    alto1, ancho1 = img1.shape
    alto2, ancho2 = img2.shape
    print(img1.shape, img2.shape)
    print(max(alto1, alto2), max(ancho1, ancho2))
    
    #Crea matrices de 0 del tamaño de la imagen mayor
    zeros_img1 = np.zeros((max(alto1, alto2), max(ancho1, ancho2)))
    zeros_img2 = np.zeros((max(alto1, alto2), max(ancho1, ancho2)))
    
    #Agraga las imagenes originales a las matrices de ceros
    zeros_img1[:img1.shape[0],:img1.shape[1]] = img1
    zeros_img2[:img2.shape[0],:img2.shape[1]] = img2
    
    if (operacion == 1):
        #AND de imagenes
        resultado = np.logical_and(zeros_img1, zeros_img2).astype(int)
        operacion = 'AND'
    elif (operacion == 2):
        #OR de imagenes
        resultado = np.logical_or(zeros_img1, zeros_img2).astype(int)
        operacion = 'OR'
    else:
        #XOR de imagenes
        resultado = np.logical_xor(zeros_img1, zeros_img2).astype(int)
        operacion = 'XOR'
        
    resultado[resultado > 0] = 255
    
    nombre_img = dir_img1.replace('.jpg','') + '-' + operacion + '-' + dir_img2.replace('.jpg','') + '.jpg'
    cv2.imwrite(nombre_img, resultado)
    
    return nombre_img, resultado
```

## Problema #3:

Desarrolle una función que dada una imagen en escala de grises muestre dicha imagen en 3 dimensiones, 
donde los ejes X e Y representan el dominio espacial de la imagen y Z representa la
intensidad del color gris de cada pixel.

```python
def representacion_3d(dir_img):
    #Importar la imagen
    img = cv2.imread(dir_img)
    
    #BGR = Azul, Verde y Rojo
    print(img.shape)
    azul = np.array(img[:, :, 0]).flatten()
    verde = np.array(img[:, :, 1]).flatten()
    rojo = np.array(img[:, :, 2]).flatten()
    
    #Transformacion a blanco y negro
    gray = np.multiply(azul, 0.114) + np.multiply(verde, 0.587) + np.multiply(rojo, 0.299)
    gray = gray.astype(np.int)
    gray = gray.reshape((img.shape[0], img.shape[1]))
        
    #Matriz de las cordenadas completas de la imagen
    valores_x, valores_y = np.mgrid[0:gray.shape[0], 0:gray.shape[1]]
    
    #Angulos para desplegar el grafico
    angulos = [90, 60, 45, 25, 0]

    #Creación de la figura y rotación
    for angle in angulos:
        fig = plt.figure()
        fig.set_size_inches(12, 7)
        ax = fig.gca(projection='3d')
        ax.plot_surface(valores_x, valores_y, gray ,rstride=1, cstride=1, cmap = plt.cm.gray, linewidth=0)
        ax.view_init(angle, 90 - angle)
        plt.show()
```

## Problema #4:
Implemente una función para cada una de las trasformaciones vistas en clase: negativa, lineal,
logarítmica y exponencial.

```python
#transformacion 1: negativa
#transformacion 2: lineal
#transformacion 3: logaritmica
#transformacion 4: exponencial

def transformaciones(dir_img, transformacion, alpha = 0.5, offset = 0, expo = 2):
    #Importar la imagen
    img = cv2.imread(dir_img)
    
    #BGR = Azul, Verde y Rojo
    print(img.shape)
    azul = np.array(img[:, :, 0]).flatten()
    verde = np.array(img[:, :, 1]).flatten()
    rojo = np.array(img[:, :, 2]).flatten()
    
    #Transformacion a blanco y negro
    gray = np.multiply(azul, 0.114) + np.multiply(verde, 0.587) + np.multiply(rojo, 0.299)
    gray = gray.astype(np.int)
    #gray = gray.reshape((img.shape[0], img.shape[1]))
    
    if (transformacion == 1):
        s = 255 - gray
        trans = "negativa"
    elif (transformacion == 2):
        s = alpha * gray + offset
        trans = "lineal"
    elif (transformacion == 3):
        s = alpha * np.log(1 + gray) 
        trans = "logaritmica"
    else:
        gray_norm = (gray / 255)
        s = 255 * (gray_norm ** expo)
        trans = "exponencial"
        
    
    s = s.astype(np.int)
    s = s.reshape((img.shape[0], img.shape[1]))
    
    nombre_img = dir_img.replace('.jpg','') + '-' + trans + '.jpg'
    cv2.imwrite(nombre_img, s)
    
    return nombre_img, s
```

```python
#Transformacion negativa
verImagen(transformaciones("antigua.jpg", 1)[0])
```

```python
#Transformacion lineal
verImagen(transformaciones("antigua.jpg", 2)[0])
```

```python
#Transformacion lineal
verImagen(transformaciones("antigua.jpg", 2, alpha = 0.8, offset = 40)[0])
```

```python
#Transformacion logaritmica
verImagen(transformaciones("antigua.jpg", 3, alpha = 40)[0])
```

```python
#Transformacion exponencial
verImagen(transformaciones("antigua.jpg", 4, expo = 0.04)[0])
```

```python
#Transformacion exponencial
verImagen(transformaciones("antigua.jpg", 4, expo = 5)[0])
```
